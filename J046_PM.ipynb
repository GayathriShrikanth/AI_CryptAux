```
import pandas as pd
df=pd.read_csv("/content/CarPrice_Assignment.csv")
```

```
X=df.drop(["price"], 1)
y=df["price"]
df.head()
```

```
df["CarName"].value_counts()#lot cat
df["fueltype"].value_counts()#Label
df["aspiration"].value_counts()#label
df["doornumber"].value_counts()#label
df["carbody"].value_counts()#1hot
df["drivewheel"].value_counts()#1hot
df["enginelocation"].value_counts()#label
df["enginetype"].value_counts()#7categories
df["cylindernumber"].value_counts()#convert no or leave
df["fuelsystem"].value_counts()#8 cat 
```

```
df.dtypes
```

```
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X['fueltype']= le.fit_transform(df['fueltype'])
X['aspiration']= le.fit_transform(df['aspiration'])
X['doornumber']= le.fit_transform(df['doornumber'])
X['enginelocation']= le.fit_transform(df['enginelocation'])
```

```
X=X.drop(["car_ID"], 1)
```

```
#One hot encoding the columns
X = pd.get_dummies(X,columns=["CarName","carbody", "drivewheel", "enginetype", "cylindernumber", "fuelsystem"])
```

```
X.head()
```

```
#Checking for null values/missing values
df.isna().sum()
```

```
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) 
```

```
#Performing linear regression
linReg=LinearRegression()
linReg.fit(X_train,y_train)
y_pred=linReg.predict(X_test)
score=linReg.score(X_test,y_test)
print(score)
print ("R2 = ",r2_score(y_test, y_pred))
#Accuracy on test
```

```
#Accuracy on train
linReg.score(X_train,y_train)
```
